# 3. Задание (на листочке) - Решите задачу:


Площадь пола прямоугольной комнаты равна 48 м2,а его периметр равен 28 м. 
Найдите длину и ширину комнаты.


```python
2x+2y=28
x*y=48
=> y=48/x =>2x+2*48/x=28
```

```python
Disc=14**2-4*1*48
Disc
```

```python
import math as mh
x1=(14+mh.sqrt(Disc))/2
x2=(14-mh.sqrt(Disc))/2
x1,x2
```

```python
y1=48/x1
y2=48/x2
y1,y2
```

Ответ комната 6x8
